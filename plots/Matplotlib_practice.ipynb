```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
import numpy as np
```

```
x = np.linspace(0,5,11)
y = x**2
print(x)
```

# Functional

```
plt.plot(x,y)
plt.xlabel('Yo boi')
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# OOP

```
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'red')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('YO BOI')
```

```
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'red')
axes2.plot(y,x,'blue')
```

```
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(8,4))
axes[0].plot(x,y,'red')
axes[1].plot(y,x,'blue')
plt.tight_layout()
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color='red',label='X squared', lw=2, alpha=0.5,linestyle='--')
ax.plot(x,x**3,color='#ff8c00',label='X cubed', lw=2, alpha=0.8,linestyle='-.',marker='x',markersize=12)
ax.legend(loc=0)
```

```
plt.scatter(x,y,color='orange')
```
